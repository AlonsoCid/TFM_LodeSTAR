## Create the base tensor

```python
import torch
import matplotlib.pyplot as plt
import torch.nn.functional as F

# Initialize a 40x40 tensor filled with 0s
tensor = torch.zeros((40, 40))

# Calculate the square
square_size = 10

start_idx = (40 - square_size) // 2
end_idx = start_idx + square_size
tensor[start_idx:end_idx, start_idx:end_idx] = 1

plt.imshow(tensor, cmap='gray', origin='lower')
plt.tick_params(axis='both', which='major', labelsize=12)
plt.title('1-.sum()/.numel() =', fontsize=20)
```

## Big mask

```python
tensor2 = tensor.unsqueeze(0).unsqueeze(0)  # Shape becomes [1, 1, 40, 40]

kernel = torch.ones((1, 1, 3, 3))
dilated_mask = F.conv2d(tensor2, kernel, padding=1) > 0
overestimation_loss = 0.5*(dilated_mask.float() - tensor2).mean()

print(overestimation_loss)

plt.imshow(dilated_mask[0][0], cmap='gray', origin='lower')
plt.tick_params(axis='both', which='major', labelsize=12)
plt.title('Big_mask - Mask = 0.01', fontsize=20)
```

## Dropout2D example

```python
import torch.nn as nn

dropout = nn.Dropout2d(p=0)
tensor_dropout = dropout(tensor)

plt.imshow(tensor_dropout, cmap='gray', origin='lower')
plt.tick_params(axis='both', which='major', labelsize=12)
plt.title('DropoutD2 p=0.01', fontsize=20)
```
